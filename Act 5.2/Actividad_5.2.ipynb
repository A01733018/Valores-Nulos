```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_excel('Gastos y costos 20-23.xlsx')
data
```

```python
#Eliminar primera fila o filas innecesarias
data= pd.read_excel('Gastos y costos 20-23.xlsx', skiprows=5)
data
```

```python
#Realizamos una copia del dataframe
data1=data.copy()
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python

#Sustituir valores nulos por un valor numÃ©rico en  concreto 
data1["FOLIO"]= data1["FOLIO"].fillna(0) 
data1.head()

#Aqui busco dejar una marca numerica para saber que folios faltan ya que este es unico para las personas en todos los casos.
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Act5.2_FOLIO.csv")
```

```python
#Sustituir valores nulos por un string en  concreto 
data1["GASTO"] =data1["GASTO"].fillna("NO EXISTE") 

#Esta columna contiene palabras por lo que para identificar los faltantes agregue la palabra "no existe"
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Act5.2_GASTO.csv")
```

```python
#Sustituir valores nulos por un valor numÃ©rico en  concreto 
data1["TC"]= data1["TC"].fillna(0) 
data1.head

#En su mayoria de esa columna estan escritos un 1 o una secuencia de numeros por lo que para identificar y eliminar los valores nulos agregue el numero 0.
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Act5.2_TC.csv")
```

```python

#Sustituir valores nulos con promedio o media
data1["IMPORTE"]=data1["IMPORTE"].fillna(round(data["IMPORTE"].mean(),1))

#Al ser resultados numericos creo lo mÃ¡s sensato es hacer el promedio de todos los resultados para que no afecten de una forma muy brusca la informaciÃ³n. 
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Act5.2_IMPORTE.csv")
```

```python
#Sustituir valores nulos con mediana
data1["IVA"]=data1["IVA"].fillna(round(data["IVA"].median(),1))

#Todos los valores son muy variados por lo que no se puede saber un estimado, use la media ya que queria usar un nuevo metodo que no aleje tanto los resultados de la realidad.
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Act5.2_IVA.csv")
```
